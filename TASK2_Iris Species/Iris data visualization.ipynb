```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_data = pd.read_csv('iris.csv')
```

```python
iris_data.head()
```

Create a histogram by counting the occurrences of each species in the dataset and plotting the counts as bars using `plt.bar()`. The x-axis represents the species, and the y-axis represents the count. 

```python
# Extract the target variable (Species)
species = iris_data["Species"]

# Count the occurrences of each species
species_counts = species.value_counts()

# Plot the histogram
plt.bar(species_counts.index, species_counts.values,color="#7752FE")

# Set labels and title
plt.xlabel("Species",color="#27005D")
plt.ylabel("Count",color="#27005D")
plt.title("Distribution of Iris Species", fontweight="bold")

# Show the plot
plt.show()
```

### Visualize the distribution of Numerical features

```python
# Define a custom color palette for the species
custom_palette = ["dodgerblue", "darkorange", "forestgreen"]

# Set the style of the plots
sns.set(style="darkgrid", palette=custom_palette)

# Create subplots for each variable
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Create histograms for Sepal Length
sns.histplot(data=iris_data, x="SepalLengthCm", hue="Species", binwidth=0.2, ax=axes[0, 0]
).set(xlabel="Sepal Length (cm)", ylabel="Frequency")
axes[0, 0].axvline(x=iris_data["SepalLengthCm"].mean(), linestyle="dashed", color="grey")

# Create histograms for Sepal Width
sns.histplot(data=iris_data, x="SepalWidthCm", hue="Species", binwidth=0.2, ax=axes[0, 1]
).set(xlabel="Sepal Width (cm)", ylabel="Frequency")
axes[0, 1].axvline(x=iris_data["SepalWidthCm"].mean(), linestyle="dashed", color="grey")

# Create histograms for Petal Length
sns.histplot(data=iris_data, x="PetalLengthCm", hue="Species", binwidth=0.2, ax=axes[1, 0]
).set(xlabel="Petal Length (cm)", ylabel="Frequency")
axes[1, 0].axvline(x=iris_data["PetalLengthCm"].mean(), linestyle="dashed", color="grey")

# Create histograms for Petal Width
sns.histplot(data=iris_data, x="PetalWidthCm", hue="Species", binwidth=0.2, ax=axes[1, 1]
).set(xlabel="Petal Width (cm)", ylabel="Frequency")
axes[1, 1].axvline(x=iris_data["PetalWidthCm"].mean(), linestyle="dashed", color="grey")

# Set the title
fig.suptitle("Iris Frequency Histogram", fontsize=15)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

## Iris Density Plot

The density plots shown below provide a visualization of the attribute distributions in the Iris dataset. Each attribute, including Petal Length, Petal Width, Sepal Width, and Sepal Length, is analyzed for the three different Iris species: Setosa, Versicolor, and Virginica. These density plots offer insights into the data distribution and class separability for each attributeal Length for each class.

These density plots provide valuable information about the distribution and class separation for each attribute, aiding in the understanding of the Iris dataset.


```python
# Set the style of the plots
sns.set(style="darkgrid")

# Create subplots for each variable
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Create density plots for Petal Length
sns.kdeplot(data=iris_data, x="PetalLengthCm", hue="Species", fill=True, common_norm=False, ax=axes[0, 0])
axes[0, 0].axvline(x=iris_data["PetalLengthCm"].mean(), linestyle="dashed", color="grey")
axes[0, 0].set(xlabel="Petal Length (cm)", ylabel="Density")

# Create density plots for Petal Width
sns.kdeplot(data=iris_data, x="PetalWidthCm", hue="Species", fill=True, common_norm=False, ax=axes[0, 1])
axes[0, 1].axvline(x=iris_data["PetalWidthCm"].mean(), linestyle="dashed", color="grey")
axes[0, 1].set(xlabel="Petal Width (cm)", ylabel="Density")

# Create density plots for Sepal Width
sns.kdeplot(data=iris_data, x="SepalWidthCm", hue="Species", fill=True, common_norm=False, ax=axes[1, 0])
axes[1, 0].axvline(x=iris_data["SepalWidthCm"].mean(), linestyle="dashed", color="grey")
axes[1, 0].set(xlabel="Sepal Width (cm)", ylabel="Density")

# Create density plots for Sepal Length
sns.kdeplot(data=iris_data, x="SepalLengthCm", hue="Species", fill=True, common_norm=False, ax=axes[1, 1])
axes[1, 1].axvline(x=iris_data["SepalLengthCm"].mean(), linestyle="dashed", color="grey")
axes[1, 1].set(xlabel="Sepal Length (cm)", ylabel="Density")

# Set the title
fig.suptitle("Iris Density Plot", fontsize=15)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Set the style of the plots
sns.set(style="whitegrid")

# Create subplots for each variable
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Create box plots for Sepal Length
sns.boxplot(iris_data,x="Species",y="SepalLengthCm",ax=axes[0,0])
axes[0, 0].set(ylabel="Sepal Length (cm)")

# Create box plots for Sepal Width
sns.boxplot(data=iris_data, x="Species", y="SepalWidthCm", ax=axes[0, 1])
axes[0, 1].set(ylabel="Sepal Width (cm)")

# Create box plots for Petal Length
sns.boxplot(data=iris_data, x="Species", y="PetalLengthCm", ax=axes[1, 0])
axes[1, 0].set(xlabel="Species", ylabel="Petal Length (cm)")

# Create box plots for Petal Width
sns.boxplot(data=iris_data, x="Species", y="PetalWidthCm", ax=axes[1, 1])
axes[1, 1].set(xlabel="Species", ylabel="Petal Width (cm)")

# Set the title
fig.suptitle("Sepal and Petal Box Plot", fontsize=15)

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
